```python
import json

db_file = "koso-tasks-embeddings.sqlite"
file_path = "koso-dogfood-export-2025-7-5-12-2.json"

with open(file_path, "r") as f:
    data = json.load(f)
project_id = data.get("projectId")
tasks = data["graph"]

tasks
```

```python
await dedupe.compute_embeddings(db_file, tasks)
```

```python
dedupe.compute_clusters(db_file, tasks)
```

```python
dupes = dedupe.compute_dupes(project_id, db_file, tasks)
```

```python
dedupe.store_dupes(dupes, project_id)
```
